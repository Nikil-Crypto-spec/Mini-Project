### Part 4: Visualization using Matplotlib

This part introduces Matplotlib, a foundational library for creating static plots in Python. It showcases different chart types to visualize the Iris data.

*   **`import matplotlib.pyplot as plt`**: Imports Matplotlib's plotting module.
*   **Line Chart (`plt.plot`)**: Displays `sepal_length` against the DataFrame index, showing how the values change sequentially.
*   **Bar Chart (`plt.bar`)**: Visualizes the counts of each `species`, providing a clear comparison of their frequencies.
*   **Scatter Plot (`plt.scatter`)**: Shows the relationship between `sepal_length` and `petal_length`, helping to identify patterns or correlations.
*   **Histogram (`plt.hist`)**: Illustrates the distribution of `sepal_width`, revealing how frequently different sepal width values occur.

```python
#PART 4: Visualization using Matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

This cell imports the `pandas` library for data manipulation and the `matplotlib.pyplot` module, aliased as `plt`, which is the primary plotting interface for Matplotlib. Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
#Task 1:
#CHART 1: Line Chart
plt.figure(figsize = (10,8))
plt.plot(df.index, df["sepal_length"], color="Skyblue")
plt.title("Sepal Length by index")
plt.xlabel("Index")
plt.ylabel("Sepal_length")
plt.show()
plt.close()
```

This cell generates a line chart of 'sepal_length' against the DataFrame index. It sets the figure size, adds a title and labels for the axes, and then displays the plot. A line chart is suitable for showing trends over an ordered sequence, though here the index order is arbitrary relative to sepal length.

```python
#CHART 2: Bar Chart
plt.figure(figsize = (10,8))
Species_counts = df['species'].value_counts()
plt.bar(Species_counts.index,Species_counts.values)
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
plt.close()
```

This cell creates a bar chart showing the count of each species. It first calculates the value counts of the 'species' column, then uses `plt.bar()` to plot these counts. Bar charts are excellent for comparing discrete categories.

```python
#CHART 3: Scatter Plot
plt.figure(figsize = (10,8))
plt.scatter(df['sepal_length'], df['petal_length'],marker ='o', color='Darkblue')
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
plt.close()
```

This cell generates a scatter plot comparing 'sepal_length' and 'petal_length'. Each point on the plot represents an iris flower, with its sepal length on the x-axis and petal length on the y-axis. Scatter plots are great for visualizing the relationship between two numerical variables.

```python
#CHART 4 :Histogram
plt.figure(figsize =(10,8))
plt.hist(df['sepal_width'],color="Darkblue")
nbins = 20
plt.xlabel("Sepal Width")
plt.ylabel("Frequency")
plt.show()
plt.close()
```

This cell creates a histogram of the 'sepal_width' column. A histogram shows the distribution of a numerical variable by dividing its range into bins and counting how many values fall into each bin. This helps in understanding the frequency of different sepal width values.

### Part 5: Visualization using Plotly

This section demonstrates Plotly Express, a high-level API for creating interactive and aesthetically pleasing visualizations. Unlike Matplotlib, Plotly charts are interactive by default.

*   **`import plotly.express as px`**: Imports the Plotly Express library.
*   **Interactive Scatter Plot (`px.scatter`)**: Similar to Matplotlib's scatter plot, but with added interactivity (zoom, pan, hover) and coloring points by `species`.
*   **Interactive Box Plot (`px.box`)**: Visualizes the distribution of `petal_length` for each `species`, showing median, quartiles, and outliers.
*   **Interactive Histogram (`px.histogram`)**: An interactive version of the `sepal_length` histogram.
*   **Interactive Pie Chart (`px.pie`)**: Displays the proportion of each `species` in the dataset, useful for categorical distribution.

```python
#Part 5: Visualization using Plotly:
import pandas as pd
import plotly.express as px
```

This cell re-imports `pandas` and `plotly.express`. Although already imported at the beginning of the notebook, explicitly importing again ensures these libraries are available for this section, especially if the kernel was reset or executed out of order.

```python
iris = px.data.iris()
```

This cell re-loads the Iris dataset into a DataFrame named `iris`. This is a common practice to ensure a clean slate for visualization, especially when working with different visualization libraries or parts of the notebook.

```python
#CHART 1: Interactive Scatter Plot
fig = px.scatter(
    iris,
    x="sepal_length",
    y="petal_length",
    color="species",
    title="sepal_length scatter vs petal_length scatter"
)

fig.show()
```

This cell generates an interactive scatter plot using Plotly Express. It plots 'sepal_length' against 'petal_length', coloring the points by 'species'. Plotly charts are interactive, allowing zooming, panning, and hovering to see data details.

```python
#CHART 2: Interactive Box Plot
fig = px.box(
    iris,
    x="species",
    y="petal_length",
    color="species",
    title="species vs petal_length Box Plot"
)

fig.show()
```

This cell creates an interactive box plot using Plotly Express. It shows the distribution of 'petal_length' for each 'species'. Box plots are useful for visualizing the spread and central tendency of data, and identifying potential outliers, across different categories.

```python
#CHART 3: Interactive Histogram
fig = px.histogram(
    iris,
    x ="sepal_length",
    nbins=20,
    title="sepal_length Histogram"
)

fig.show()
```

This cell generates an interactive histogram of 'sepal_length' with 20 bins. Similar to Matplotlib's histogram, it visualizes the distribution of sepal length, but with Plotly's interactivity features.

```python
#CHART 4 :Interactive Pie Chart
Species_count=iris["species"].value_counts().reset_index()
Species_count.columns=["species","count"]
fig = px.pie(
    Species_count,
    names="species",
    values="count",
    title="Pie Chart for species count"

)

fig.show()
```

This cell creates an interactive pie chart to show the proportion of each species in the dataset. It first calculates the counts of each species, then uses `px.pie()` to visualize these proportions. Pie charts are effective for showing parts of a whole.

### Part 6: Python Fundamentals

This final section revisits basic Python list operations, demonstrating how to create, modify, and analyze lists.

*   **`marks = [78, 82, 91, 65, 88]`**: Creates a simple Python list of integers.
*   **`marks.append(90)`**: Adds a new item to the end of the list.
*   **`marks.remove(91)`**: Removes a specific item by its value from the list.
*   **`marks.sort()`**: Sorts the list in ascending order *in-place*.
*   **`max(marks)` and `min(marks)`**: Finds the highest and lowest values in the list, respectively.

In summary, this notebook provides a comprehensive introduction to essential data science tools and Python concepts, from data loading and initial exploration to statistical analysis and various visualization techniques.
